Given num a 4-digit number, we want to find two integers new1 and new2 created using digits of num where new1+new2 is the minimum possible

```python
def minimumSum(num):
        #we turn num into a string so we could take each digit later
        num=str(num)
        #we store the different digits in a list calles digits after converting them to integers
        digits=[int(num[0]),int(num[1]),int(num[2]),int(num[3])]
        #we sort the list in the ascending order
        digits.sort()
        #the least small numbers will be dozens and the two that left will be the units
        new1=int(str(digits[0])+str(digits[2]))
        new2=int(str(digits[1])+str(digits[3]))
        return new1+new2
```

Example 1:

```python
num = 2932
#expected output: 52
minimumSum(num)
```

Example 2:

```python
num = 4009
#expected output: 13
minimumSum(num)
```
